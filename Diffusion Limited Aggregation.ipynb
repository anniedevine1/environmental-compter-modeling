```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FFMpegWriter
metadata = dict(title = "Pipe Animation",artist = "Matplotlib")
writer = FFMpegWriter(fps=15,metadata=metadata, bitrate=200000)
```

```python
#note, this function expects a matrix A[ix,iy] 
#and then displays so that A[:,0] is the lowest row of pixels
def display(A):
    maxX = A.shape[0]
    maxY = A.shape[1]
    B = np.zeros((maxY, maxX))
    for ix in range(0,maxX):
        for iy in range(0,maxY):
            B[maxY-1-iy,ix] = A[ix,iy]

    #Display the graphics outside of the notebook. 
    #On a PC, use '%matplotlib qt' instead.
    %matplotlib tk 
    
    plt.rcParams['figure.figsize'] = [6, 6/maxX*maxY]
    plt.imshow(B); 
    plt.axis('off'); 
    plt.show()
    plt.draw()
    plt.pause(0.01)
```

```python
nParticles = 20000
maxX = 200
maxY = 40
```

```python
# Initialize matrix containing all 2D grid points A(x,y)
# 0 <= x < maxX
# 0 <= y < maxY
# A(x,y)=0 ... site is empty
# A(x,y)>0 ... site is filled
A = np.zeros((maxX, maxY))

# Introduce a sticky wall at the bottom 
# by filling the lowest row of pixels with particles
A[:,0] = 1
A[:,maxY-1] = 1
print(A.transpose())
```

```python
succesful_particles = 0
total_steps = 0
max_steps_per_particle = maxX*maxY*2
frames = []
for i in range(0,nParticles):
    # Compute new starting point on the line y=yStart
    x = maxX - 1
    y = maxY//2; #always start at upper limit
    current_steps = 0
    attached = False

    while not attached and current_steps < max_steps_per_particle:
        current_steps += 1
        x_org,y_org = x,y

        r = np.random.random(); # Random float:  0.0 <= r < 1.0
        #based on the value of 'r', move the particle
        #left, right, up, or down and change x and y accordingly
        ...
        if(r < 0.1):
            x += 1
        elif(r < 0.6):
            x -= 1
        elif(r < 0.8):
            y += 1
        else:
            y -= 1
        #boundaries
        if x >= maxX:
            x = x_org
            continue
        if x<0:
            succesful_particles += 1
            total_steps += current_steps
            break
        
        if y<0 or y>= 39: 
            total_steps += current_steps
            break
        if A[x,y] == 1:
            total_steps += current_steps
            break
display(A)
probability = succesful_particles/nParticles
print("The number of particles that made it to the end is", succesful_particles)
print("The probability that a particle reaches the end of the pip is", probability*100, "percent")
```

```python
nFrames = 50
B=np.zeros((nFrames,maxX,maxY))

def display(A,B):
    maxX = A.shape[0]
    maxY = A.shape[1]

    m2 = np.max(B)
    B[A[:,:] == 1] = m2
    if (m2 > 0):
        m1 = np.min(B[B[:,:]>0])
        if (m1==m2): m1=m2/2
        B[B[:,:]==0] = m1*(m1/m2)**(1/10)

    D = np.zeros((maxY, maxX))
    for iy in range(0,maxY):
        D[maxY-1-iy,:] = np.log(B[:,iy])
    
    plt.imshow(D);
    plt.axis('off');
    plt.show()
    plt.draw()
    plt.pause(0.1)
```

```python
succesful_particles = 0
total_steps = 0
max_steps_per_particle = maxX*maxY*2
nFrames=50
record_step = 10
frame_index = 0

for i in range(0,nParticles):
    # Compute new starting point on the line y=yStart
    x = maxX - 1
    y = maxY//2; #always start at upper limit
    current_steps = 0
    attached = False

    while not attached and current_steps < max_steps_per_particle:
        current_steps += 1
        x_org,y_org = x,y

        r = np.random.random(); # Random float:  0.0 <= r < 1.0
        #based on the value of 'r', move the particle
        #left, right, up, or down and change x and y accordingly
        ...
        if(r < 0.1):
            x += 1
        elif(r < 0.6):
            x -= 1
        elif(r < 0.8):
            y += 1
        else:
            y -= 1
        #boundaries
        if x >= maxX:
            x = x_org
            continue
        if x<0:
            succesful_particles += 1
            total_steps += current_steps
            break
        
        if y<=1 or y>= 38: 
            total_steps += current_steps
            break
        if A[x,y] == 1:
            total_steps += current_steps
            break

        if current_steps % record_step == 0 and frame_index < nFrames:
            B[frame_index,x,y] += 1
            frame_index +=1
probability = succesful_particles/nParticles
print("The number of particles that made it to the end is", succesful_particles)
print("The probability that a particle reaches the end of the pip is", probability*100, "percent")

for f in range(nFrames):
    display(A,B[f,:,:])
```

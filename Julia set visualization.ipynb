```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Julia(z,C):
    z_new = z**2 + C
    return z_new
```

```python
n = 201
data = np.zeros((n,n))
x_values = np.linspace(-3,3,201)
y_values = np.linspace(-3,3,201)
N = len(x_values)
```

```python
for i in range(N):
    for j in range(N):
        x_val = x_values[i]
        y_val = y_values[j]
        C = complex(-0.4,+0.6)
        z = complex(x_val,y_val)
        k = 0
        max_iterations = 50
        cutoff = 50
        for k in range(1,max_iterations + 1):
            z = Julia(z,C)
            if abs(z)>= cutoff:
                break
        if abs(z) >= cutoff:
            data[i,j] = k
        else:
            data[i,j] = max_iterations
```

```python
plt.imshow(data)
```

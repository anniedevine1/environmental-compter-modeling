```python
%matplotlib tk
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib.animation import FFMpegWriter
metadata = dict(title='My first animation in 2D', artist='Matplotlib',comment='Wakanda is coming.')
writer = FFMpegWriter(fps=15, metadata=metadata)
fig = plt.figure()

with writer.saving(fig, "lab7_part1.mp4", dpi=200):
    N = 50
    L = 30
    T = np.zeros(N+1)
    T[0] = 100
    T[L] = 0
    x = np.linspace(0,L,N+1)

    #iterations
    n = 1000
    T_new = np.copy(T)
    for j in range(n):
        for i in range(1,N):
            T_new[i] = (0.5)*((T_new[i-1])+(T_new[i+1]))
        plt.clf()
        plt.plot(x,T_new, 'ro-',mfc='w')
        plt.show()
        plt.draw()
        plt.pause(0.05)
```

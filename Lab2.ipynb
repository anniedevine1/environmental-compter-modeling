```python
#This code generates random numbers and rounds them to either zero or 1. It then lays out these numbers in an 11 by 11 grid and gives each number a square. It then fills the square with a color depending on if the number is 0 or 1.
```

```python
n = 11
data = np.zeros((n,n))
# (a) what is the effect of the following three lines
data[1,1] = 1
data[2,3] = 1
data[3,5] = 1
# (b) again what do the following two lines do?
data[5,1:n-1] = 1
data[n-2,:] = 1
print(data)
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
#a) these lines change the zeros to ones in the designated spaces, eg. the row with value 1 and column with value 1.
#b) these lines change multiple zeros to ones at once, but indicating which row and column spaces to change.
#finally the numbers are turned into colored squares based on what number they are.
```

```python
# Here we practise 'for' loops
n = 11
data = np.zeros((n,n))
# (a) What is the effect of this 'for' loop?
for i in range(0,n):
    data[i,3] = 1
# (b) And this 'for' loop?
col = (n-1) // 2
for i in range(0,n,2):
    data[i,col] = 1
print(data)
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
#a) the effect of the for loop is, it changes every the value of the third column to 1 for every row (0,11).
```

```python
#b) this for loop replaces every other row entry in column 5 with a 1.
```

```python
n = 11
data = np.zeros((n,n))
for i in range (0,n):
    data[i,i] = 1

for i in range (n-1,-1,-1):
    data[i,10-i] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```

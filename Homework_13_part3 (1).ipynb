```python
import numpy as np
import datetime
```

```python
def function(x):
    f = np.exp(-x**2)
    f = np.sin(f+2.0)
    return np.sqrt(f+10.0)
```

```python
n = 2000000
x = np.random.random(n)
f = np.zeros(n)

start = datetime.datetime.now()
for i in range(n):
    f[i] = function(x[i])
now = datetime.datetime.now()
print(f)
print ('Execution time (hours:minutes:seconds) = ',str(now-start))
```

```python
start = datetime.datetime.now()
f2 = function(x)
now = datetime.datetime.now()
print(f2)
print ('Execution time (hours:minutes:seconds) = ',str(now-start))
```

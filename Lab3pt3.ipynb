```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import math
%config InlineBackend.figure_format = 'retina'
```

```python
files_2018 = glob.glob('submissions_2018/*.txt')
files_2019 = glob.glob('submissions_2019/*.txt')
files_2020 = glob.glob('submissions_2020/*.txt')
files_2021 = glob.glob('submissions_2021/*.txt')
files_2022 = glob.glob('submissions_2022/*.txt')
files_2023 = glob.glob('submissions_2023/*.txt')
```

```python
positions_2018 = np.loadtxt('x_positions_2018.txt')
positions_2019 = np.loadtxt('x_positions_2019.txt')
positions_2020 = np.loadtxt('x_positions_2020.txt')
positions_2021 = np.loadtxt('x_positions_2021.txt')
positions_2022 = np.loadtxt('x_positions_2022.txt')
positions_2023 = np.loadtxt('x_positions_2023.txt')
```

```python
numbers_2018 = len(positions_2018)
numbers_2019 = len(positions_2019)
numbers_2020 = len(positions_2020)
numbers_2021 = len(positions_2021)
numbers_2022 = len(positions_2022)
numbers_2023 = len(positions_2023)
```

```python
stdDev_2018 = []
for file in files_2018:
    data = np.loadtxt(file)
    standard_deviation_2018 = math.sqrt(sum((data-positions_2018)**2)/numbers_2018)
    print(standard_deviation_2018)
    stdDev_2018.append(standard_deviation_2018)
```

```python
x_indexes_2018 = np.arange(0,len(stdDev_2018),1)
ranked_stdDev_2018 = np.sort(stdDev_2018)
plt.plot(x_indexes_2018, ranked_stdDev_2018)
```

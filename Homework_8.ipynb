```python
#1
k = 0.042
rho = 2.55
cp = 1.23
t_lava = 1150
t_rock = 20
N_dike = 50
N_total = 5*N_dike
length_dike = 500
length_total = 5*length_dike
dx = length_total/N_total
x = np.linspace(0,250,N_total + 1)
dt = 240 

T = np.full(N_total + 1, t_rock)
T[100:150] = t_lava
T_new = np.copy(T)
plt.clf()
plt.plot(x,T, 'ro-',mfc='w')
plt.show()
plt.draw()
plt.pause(0.05)
```

```python
#I am using centimeters in this code and my unit of length
#I am using seconds for my unit of time
#The coefficient k is a heat conductance of 0.042 J/cm/seconds/Kelvin.
#My delta x in this grid is 10 centimeters
#My delta t is 240 seconds, or 4 minutes
#the coefficient is (dt*k)/(dx^2*cp*rho), which all has to be less than 1/8
```

```python
#2
#finding converging results
N_values = [50,100,150,200,250,300]
results = {}
for N_dike in N_values:
    k = 0.042
    rho = 2.55
    cp = 1.23
    t_lava = 1150
    t_rock = 20
    N_total = 5*N_dike
    length_dike = 500
    length_total = 5*length_dike
    dx = length_total/N_total
    x = np.linspace(0,N_total,N_total + 1)
    eta = 0.1
    time_seconds = {
        "1 hour":3600,
        "1 day":86400,
        "1 week":604800,
        "1 month":2592000,
        "1 year":31536000
    }
    dt = (eta*(dx**2)*cp*rho)/k
    T = np.full(N_total + 1, t_rock)
    center_index = N_total//2
    elapsed_time = 0.0
    T[N_dike*2:N_dike*3+1] = t_lava
    T_new = np.copy(T)
    print(f"\nN_dike = {N_dike}, dx = {dx:.2f} cm, dt = {dt:.2f} s")
    current_N_results = {}
    sorted_time_markers = sorted(time_seconds.items(), key=lambda item: item[1])
    for time_key, target_time in sorted_time_markers:
        steps_needed = int(np.ceil((target_time - elapsed_time)/dt))
        for step in range(steps_needed):
            T_new[1:-1] = T[1:-1] + eta*(T[:-2] + T[2:] - 2*T[1:-1])
            T = T_new.copy()
        elapsed_time = target_time
        center_temp = T[center_index]
        current_N_results[time_key] = center_temp
    results[N_dike] = current_N_results
    print(f"  Center Index: {center_index}")
    for time_key, temp in current_N_results.items():
        print(f"  {time_key}: {temp:.2f}°C")
N_converged = 300
print(f"\n--- Converged Results for N={N_converged} ---")
for time_key, temp in results[N_converged].items():
    print(f"{time_key}: {temp:.2f}°C")
```

```python
#4
k = 0.042
rho = 2.55
cp = 1.23
t_lava = 1150
t_rock = 20
N_dike = 50
N_total = 5*N_dike
length_dike = 500
length_total = 5*length_dike
dx = length_total/N_total
x = np.linspace(0,N_total,N_total + 1)
eta = 0.1
time_seconds = {
    "1 hour":3600,
    "1 day":86400,
    "1 week":604800,
    "1 month":2592000,
    "1 year":31536000
}
dt = (eta*(dx**2)*cp*rho)/k
T = np.full(N_total + 1, t_rock)
center_index = N_total//2
elapsed_time = 0.0
T[N_dike*2:N_dike*3+1] = t_lava
E_initial = np.sum(T[N_dike*2 : N_dike*3 + 1] - t_rock)
T_new = np.copy(T)
heat_loss_results = {}
sorted_time_markers = sorted(time_seconds.items(), key=lambda item: item[1])
for time_key, target_time in sorted_time_markers:
    steps_needed = int(np.ceil((target_time - elapsed_time)/dt))
    for step in range(steps_needed):
        T_new[1:-1] = T[1:-1] + eta*(T[:-2] + T[2:] - 2*T[1:-1])
        T = T_new.copy()
    E_current = np.sum(T[N_dike*2 : N_dike*3 + 1] - t_rock)
    fraction_lost = 1.0 - (E_current / E_initial)
    
    heat_loss_results[time_key] = {
        "Center Temp": T[center_index],
        "Fraction Lost": fraction_lost
    }    
print(f"\n--- Heat Loss and Convergence Results for N={N_dike} ---")
print("{:<10} {:>12} {:>15}".format("Time Span", "Center Temp (°C)", "Fraction Lost (%)"))
print("-" * 37)

for time_key, data in heat_loss_results.items():
    print("{:<10} {:>12.2f} {:>15.2f}".format(
        time_key, 
        data["Center Temp"], 
        data["Fraction Lost"] * 100
    ))
```

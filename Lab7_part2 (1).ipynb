```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib tk
```

```python
from matplotlib.animation import FFMpegWriter
metadata = dict(title='My first animation in 3D', artist='Matplotlib',comment='Wakanda is here now.')
writer = FFMpegWriter(fps=15, metadata=metadata,bitrate=200000)
fig = plt.figure(dpi=200)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
with writer.saving(fig, "Lab7_part2.mp4", dpi=200):
    N = 51
    L = 30
    X = np.linspace(0, L, N+1) # where n is number of segments
    Y = np.linspace(0, L, N+1)
    X, Y = np.meshgrid(X, Y)
    fig = plt.figure(dpi=200)
    T = np.zeros((N+1,N+1))
    T[:,0] = 100
    #iterations
    nr = 1000
    T_new = np.copy(T)
    for r in range(nr):
        for i in range(1,N):
            for j in range(1,N):
                T_new[i,j] = (0.5)*(T_new[i-1,j-1]+T_new[i+1,j+1])
        fig.clear()
        ax = fig.add_subplot(111,projection='3d')
        ax.plot_surface(X, Y, T_new, cmap=cm.coolwarm, antialiased=False)
        plt.draw()
        plt.pause(0.05) # choose the time argument between 0.01 and 0.5
```

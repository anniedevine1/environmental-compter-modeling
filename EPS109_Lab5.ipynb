```python
n = 201
data = np.zeros((n,n))
```

```python
x_values = np.linspace(-2,0.5,201)
y_values = np.linspace(-1.25,1.25,201)
N = len(x_values)
```

```python
for i in range(N):
    for j in range(N):
        z = 0
        x_val = x_values[i]
        y_val = y_values[j]
        C = complex(x_val,y_val)
        k = 0
        max_iterations = 50
        cutoff = 50
        for k in range(1,max_iterations + 1):
            z = Mandelbrot(z,C)
            if abs(z)>= cutoff:
                break
        if abs(z) >= cutoff:
            data[i,j] = 1
        else:
            data[i,j] = 0
```

```python
plt.imshow(data)
```

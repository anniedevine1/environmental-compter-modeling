```python
import matplotlib.pyplot as plt
import numpy as np
import math
n = 11
data = np.zeros((n,n))
for i in range (0,n):
    data[i:n,i] = 1


plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 21
data = np.zeros((n,n))
for i in range (0,n):
    data[i:n,10] = 1
for i in range (0,n):
    data[10,i:n] = 1
for i in range (0,10):
    data[i,10-i] = 1
for i in range (0,10):
    data[i,i+10] = 1
for i in range (10,n):
    data[i,i-10] = 1

for i in range (10,n):
    data[i,9-i] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 50
data = np.zeros((n,n))
data[0,0] = 1
for i in range (0,25):
    data[0:i,i] = 1
for i in range (25,n):
    data[0:n-i,i] = 1
for i in range (25,n):
    data[i:n,i] = 1
for i in range (25,n):
    data[i:n,n-i] = 1

data[n-1,0] = 1

plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))
for i in range(0,n,2):
    for j in range(0,n,2):
        data[i,j] = 1

for i in range(1,n,2):
    for j in range(1,n,2):
        data[i,j] = 1

plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 13
data = np.zeros((n,n))
for i in range(0,n,3):
    for j in range(0,n,3):
        data[i,j] = 1

for i in range(1,n,3):
    for j in range(1,n,3):
        data[i,j] = 1

for i in range(2,n,3):
    for j in range(2,n,3):
        data[i,j] = 1
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 500
data = np.zeros((n,n))
for i in range(0,250):
    for j in range(0,250):
        if math.sqrt(i**2 + j**2) <= 250:
            data[i,j] = 1

for i in range(0,250):
    for j in range(250,n):
        if math.sqrt(i**2 + (500-j)**2) <= 250:
            data[i,j] = 1

for i in range(250,n):
    for j in range(0,250):
        if math.sqrt((500-i)**2 + j**2) <= 250:
            data[i,j] = 1

for i in range(250,n):
    for j in range(250,n):
        if math.sqrt((500-i)**2 + (500-j)**2) <= 250:
            data[i,j] = 1

plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 500
data = np.zeros((n,n))
for ix in range (0,350):
    for iy in range(75,425):
        if math.sqrt((175-ix)**2 + (250-iy)**2) <= 175:
            data[iy,ix] = 1

for ix1 in range (150,n):
    for iy1 in range (75,425):
        if math.sqrt((325-ix1)**2 + (250-iy1)**2) <= 175:
            data[iy1,ix1] = 1

for i in range (0,n):
    for j in range(74,425):
        if math.sqrt((175-i)**2 + (250-j)**2) <= 175 and math.sqrt((325-i)**2 + (250-j)**2) <= 175:
            data[j,i] = 0
    
plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
H1 = np.copy(H)
H1[45:70,70:110] = 10000
plt.imshow(H1, interpolation='nearest')
plt.show()
```

```python
H2 = np.copy(H)
for i in range (0,180):
    for j in range (0,160):
        H2[i,j] = H[i,j+200]
for i in range (0,180):
    for j in range (160,359):
        H2[i,j] = H[i,j-160]
plt.imshow(H2, interpolation='nearest')
plt.show()
```

```python
H3 = np.copy(H)
for ix in range(0,360):
    for iy in range(0,180):
        ix2 = round(90+(ix/360)*180)
        H3[iy,ix] = H2[iy,ix2]

plt.imshow(H3, interpolation = 'nearest')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib tk
from matplotlib.animation import FFMpegWriter
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
metadata = dict(title = "Letter Temperature Animation",artist = "Matplotlib")
writer = FFMpegWriter(fps=15,metadata=metadata, bitrate=200000)
```

```python
#create a function to define the shape of the letter N as the intial condition
```

```python
def initial_condition(N,high_temp = 100):
    T1 = np.zeros((N+1,N+1))
    middle = N//2
    thickness = N//10
    T1[thickness:N-thickness, thickness:2*thickness] = high_temp
    T1[thickness:N-thickness, N-2*thickness:N-thickness] = high_temp
    for i in range(thickness,N-thickness):
        xstart = thickness*2
        xend = N - thickness*2
        x = int(xend + (xstart - xend)*(i - thickness)/(N - 2*thickness))
        T1[i,x:x + thickness] = high_temp
    T1[[0,N],:] = 0
    T1[:,[0,N]] = 0
    return T1
```

```python
N = 50
L = 30
nr = 100
X = np.linspace(0, L, N+1) # where n is number of segments
Y = np.linspace(0, L, N+1)
X, Y = np.meshgrid(X, Y)
T = initial_condition(N)
fig = plt.figure(dpi=200)
```

```python
#first part
```

```python
with writer.saving(fig, "Homework_7_sim1.mp4", dpi=200):
    T_new = np.copy(T)
    for r in range(nr):
        for i in range(1,N):
            for j in range(1,N):
                T_new[i,j] = (T_new[i+1,j] + T_new[i - 1,j] + T_new[i,j + 1] + T_new[i,j - 1])*0.25
            
        fig.clear()
        ax = fig.add_subplot(111,projection='3d')
        ax.plot_surface(X, Y, T_new, cmap=cm.coolwarm, antialiased=False)
        ax.set_zlim(0,100)
        plt.draw()
        plt.pause(0.05) # choose the time argument between 0.01 and 0.5
        writer.grab_frame()
```

```python
#second part
```

```python
with writer.saving(fig, "Homework_7_sim2.mp4", dpi=200):
    sum_initial = np.sum(T)
    print("sum initial", sum_initial)
    for r in range(nr):
        T_new = np.copy(T)
        T_new[0,0] = (2*T[0,0] + T[1,0] + T[0,1])*(0.25)
        T_new[N-1,N-1] = (2*T[N-1,N-1] + T[N-2,N-1] + T[N-1,N-2])*(0.25)
        T_new[0,N-1] = (2*T[0,N-1] + T[1,N-1] + T[0,N-2])*(0.25)
        T_new[N-1,0] = (2*T[N-1,0] + T[N-2,0] + T[N-1,1])*(0.25)
        for i in range(1,N-1):
            T_new[i,0] = (T[i,0] + T[i+1,0] + T[i-1,0] + T[i,1])*(0.25)
            T_new[i,N-1] = (T[i,N-1] + T[i+1,N-1] + T[i-1,N-1] + T[i,N-2])*(0.25)

            T_new[0,i] = (T[0,i] + T[0,i+1] + T[0,i-1] + T[1,i])*(0.25)
            T_new[N-1,i] = (T[N-1,i] + T[N-1,i+1] + T[N-1,i-1] + T[N-2,i])*(0.25)
            
        for i in range(1,N-1):
            for j in range(1,N-1):
                T_new[i,j] = (T[i+1,j] + T[i - 1,j] + T[i,j + 1] + T[i,j - 1])*0.25
            
        T = np.copy(T_new)
        fig.clear()
        ax = fig.add_subplot(111,projection='3d')
        ax.plot_surface(X, Y, T, cmap=cm.coolwarm, antialiased=False)
        ax.set_zlim(0,100)
        plt.draw()
        plt.pause(0.05)
        writer.grab_frame()
    print("sum final", np.sum(T))
```

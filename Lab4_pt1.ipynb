```python
import matplotlib.pyplot as plt
import numpy as np
H = np.loadtxt('topography_180x360_grid.txt')
plt.imshow(H)
plt.show()
```

```python
for y in range(0,179):
    for x in range(0,359):
        elevation = H[y,x]
        latitude = (y-90)
        longitude = (x-180)
        if (x < 2): #This is just to keep the printing shorter
            print('latitude =', latitude, 'logitude =', longitude, 'elevation =', H[y,x])
```

```python
R = 6371
n_lat, n_lon = H.shape
dlat = np.pi/n_lat
dlon = 2*np.pi/n_lon
total_area = 0.0
for i in range(n_lat):
    lat = -np.pi/2 + (i+0.5) * dlat
    dA = R**2 * np.cos(lat) * dlat * dlon
    total_area += dA * n_lon

analytical = 4 * np.pi * R**2
print("Computed Earth surface area:", total_area, "km^2")
print("analytical value:", analytical, "km^2")
```

```python
R = 6371
n_lat, n_lon = H.shape
dlat = np.pi/n_lat
dlon = 2*np.pi/n_lon
water_area = 0.0
for i in range(n_lat):
    lat = -np.pi/2 + (i+0.5) * dlat
    for j in range(n_lon):
        if (H[i,j]<0):
            dA = R**2 * np.cos(lat) * dlat * dlon
            water_area += dA

print(water_area)
proportion_of_earth_covered_by_water = water_area/total_area
print(proportion_of_earth_covered_by_water)
```

```python
water_volume = 0.0
for i in range(n_lat):
    lat = -np.pi/2 + (i+0.5) * dlat
    for j in range(n_lon):
        if (H[i,j]< 0):
            dA = R**2 * np.cos(lat) * dlat * dlon
            water_volume += dA*abs(H[i,j])


print(water_volume)
```

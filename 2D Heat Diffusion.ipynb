```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib tk
```

```python
from matplotlib.animation import FFMpegWriter
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
metadata = dict(title = "2D Heat Diffusion",artist = "Matplotlib")
writer = FFMpegWriter(fps=15,metadata=metadata, bitrate=200000)
```

```python
N = 20
L = 1
T = np.zeros(N+1)
T[0] = 0
T[L] = 0
x = np.linspace(0,L,N+1)
for i in range(0,N):
    if i*0.05 <= 0.5:
        T[i] = 2*i*0.05
    if i*0.05 > 0.5:
        T[i] = 2*(1 - i*0.05)
plt.clf()
plt.plot(x,T, 'ro-',mfc='w')
plt.show()
plt.draw()
plt.pause(0.05)
```

```python
N = 20
L = 1
K = 1
x = np.linspace(0,L,N+1)
r = 10**-4
T_new = np.copy(T)
for i in range(1,N):
    T_new[i] = ((K*r*(N/L)**2)*(T[i+1] - 2*T[i] + T[i-1]) + T[i])
T = T_new.copy()
plt.clf()
plt.plot(x,T, 'ro-',mfc='w')
plt.show()
plt.draw()
plt.pause(0.05)
```

```python
N = 20
L = 1
K = 1
x = np.linspace(0,L,N+1)
r = 10**-4
E_start = sum(T)
E_stop = sum(T)*(0.01)
time = 0
n = 0
T_new = np.copy(T)
while sum(T) > E_stop:
    for i in range(1,N):
        T_new[i] = ((K*r*(N/L)**2)*(T[i+1] - 2*T[i] + T[i-1]) + T[i])
    T = T_new.copy()
    time += r
    n += 1
    if n % 100 ==0:  #I am only going to plot every 100 steps to same time
        plt.clf()
        plt.plot(x,T, 'ro-',mfc='w')
        plt.ylim(0,1)
        plt.show()
        plt.draw()
        plt.pause(0.05)
plt.clf()
plt.plot(x,T, 'ro-',mfc='w')
plt.ylim(0,1)
plt.show()
plt.draw()
plt.pause(0.05)
print("time taken:", time, "seconds" )
print(n, "iterations")
```

```python
L = 10
N = 40
dx = L/N
dy = dx
max_temp = 100
center = 5
slope = max_temp/center
x = np.linspace(0,L,N+1)
y = np.linspace(0,L,N+1)
X, Y = np.meshgrid(x,y)
X_distance = np.abs(X - center)
Y_distance = np.abs(Y - center)
max_distance = np.maximum(X_distance,Y_distance)
T = np.maximum(0,max_temp - max_distance*slope)
#This is creating the initial pyramid of temperatures
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, T, cmap='jet')
ax.set_zlim(0, 100)
plt.show()
```

```python
with writer.saving(fig, "2D Heat Diffusion.mp4", dpi=200):
    r = 0.01
    K = 1
    n = 0
    coefficient = K*r/(dx**2)
    T_new = np.copy(T)
    for i in range(0,2000):
        for i in range(1,N):
            for j in range(1,N):
                T_new[i,j] = (coefficient*(T[i+1,j] + T[i-1,j] + T[i,j+1] + T[i,j-1] - 4*T[i,j]) + T[i,j])
        T = T_new.copy()
        n += 1
        if n % 10 == 0:
            fig.clear()
            ax = fig.add_subplot(111,projection='3d')
            ax.plot_surface(X, Y, T, cmap='jet', antialiased=False)
            ax.set_zlim(0,100)
            plt.draw()
            plt.pause(0.05)
            writer.grab_frame()
```
